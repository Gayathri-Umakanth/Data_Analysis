```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
customer = pd.read_excel("C:/Users/My Pc/Desktop/Data_Analytics_Course/Power_BI/SuperstoreProject/Superstore_Customer_data.xlsx")
customer.head()
```

```python
customer.info()
```

```python
customer['Customer ID'] = customer['Customer ID'].astype('string')
customer['Customer Name'] = customer['Customer Name'].astype('string')
customer['Segment'] = customer['Segment'].astype('string')
customer['City'] = customer['City'].astype('string')
customer['State'] = customer['State'].astype('string')
```

```python
customer['EastWest'] = np.add(customer['East'],customer['West'])
```

```python
product = pd.read_excel("C:/Users/My Pc/Desktop/Data_Analytics_Course/Power_BI/SuperstoreProject/Superstore_Product_data.xlsx")
product.head()
```

```python
product.info()
```

```python
product['Product ID'] = product['Product ID'].astype('string')
product['Category'] = product['Category'].astype('string')
product['Sub-Category'] = product['Sub-Category'].astype('string')
product['Product Name'] = product['Product Name'].astype('string')
```

```python
product.info()
```

```python
sales = pd.read_excel("C:/Users/My Pc/Desktop/Data_Analytics_Course/Power_BI/SuperstoreProject/SuperStore_Sales_Data.xlsx", header=1)
sales.head()
```

```python
sales=sales.dropna(how='all')
```

```python
sales.head()
```

```python
sales.info()
```

```python
sales['Order ID'] = sales['Order ID'].astype('string')
sales['Product ID'] = sales['Product ID'].astype('string')
sales['Customer ID'] = sales['Customer ID'].astype('string')
sales['Order Priority'] = sales['Order Priority'].astype('string')
sales['Ship Mode'] = sales['Ship Mode'].astype('string')
sales['Payment Mode'] = sales['Payment Mode'].astype('string')
```

```python
sales.info()
```

```python
#Revenue & SaLes

plt.figure(figsize=(4,4))
sb.lineplot(x='Ship Mode',y='Sales',data=sales,marker='*',color='purple')
plt.grid(axis='y',alpha=0.3)
plt.xlabel("Shipment Mode")
plt.title("Sales based on Ship Mode")
plt.show()
```

```python
plt.figure(figsize=(4,4))
sb.barplot(data=sales, x='Order Priority', y='Sales', ci=None)
plt.title('Sales by Order Priority')
plt.ylabel('Total Sales')
plt.show()
```

```python
plt.figure(figsize=(4,4))
sb.scatterplot(data=sales, x='Revenue', y='Profit', hue='Payment Mode')
plt.title('Revenue vs. Profit')
plt.xlabel('Revenue')
plt.ylabel('Profit')
plt.show()
```

```python
sb.kdeplot(x='Sales',data=sales,fill=True,color='magenta')
plt.title('Overall Sales')
plt.xlabel('Sales')
plt.ylabel('Sales Details')
plt.show()
```

```python
sb.barplot(x='Segment',y='EastWest',data=customer,ci=None)
plt.show()
```

```python
sb.histplot(x='Quantity',data=product,bins=15,color='blue',kde=True)
plt.title('Total Number of Quantities sold')
plt.xlabel('Quantiy')
plt.ylabel('Sold Count')
plt.show()
```

```python
sales_customer = pd.merge(sales,customer, on=['Customer ID'], how='inner')
print(sales_customer)
```

```python
sales_product = pd.merge(sales,product, on=['Product ID'], how='inner')
sales_product.head()
```

```python
plt.figure(figsize=(4,4))
sb.barplot(data=sales_product, x='Category', y='Sales', ci=None)
plt.title('Sales by Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.plot(sales_product.Sales,sales_product.Profit,marker='o',color='g')
plt.title('Sales vs Profit')
plt.xlabel('Total Sales')
plt.ylabel('Total Profit')
plt.show()
```
